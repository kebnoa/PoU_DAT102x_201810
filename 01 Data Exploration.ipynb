## Read the files into Pandas

```python
features_filename = 'data/DAT102x_Predicting_Chronic_Hunger_-_Training_values.csv'
labels_filename = 'data/DAT102x_Predicting_Chronic_Hunger_-_Training_labels.csv'
final_test_filename = 'data/DAT102x_Predicting_Chronic_Hunger_-_Test_values.csv'

features = pd.read_csv(features_filename)
labels = pd.read_csv(labels_filename)
final_test = pd.read_csv(final_test_filename)

print(features.shape)
print(labels.shape)
print(final_test.shape)
```

## Explore features
Loop through all the features and draw summary plots. The plots consist of a histogram/density plot on the left, a box plot in the middle, and a box plot of the natural logarithm of the feaure to the right.

```python
train_dataset = pd.concat([features, labels], axis='columns')
features_of_interest = ['agricultural_land_area',
                        'percentage_of_arable_land_equipped_for_irrigation', 'cereal_yield',
                        'droughts_floods_extreme_temps', 'forest_area', 'total_land_area',
                        'fertility_rate', 'life_expectancy', 'rural_population',
                        'total_population', 'urban_population', 'population_growth',
                        'avg_value_of_food_production', 'cereal_import_dependency_ratio',
                        'food_imports_as_share_of_merch_exports',
                        'gross_domestic_product_per_capita_ppp',
                        'imports_of_goods_and_services', 'inequality_index',
                        'net_oda_received_percent_gni', 'net_oda_received_per_capita',
                        'tax_revenue_share_gdp', 'trade_in_services',
                        'per_capita_food_production_variability',
                        'per_capita_food_supply_variability', 'adult_literacy_rate',
                        'school_enrollment_rate_female', 'school_enrollment_rate_total',
                        'avg_supply_of_protein_of_animal_origin',
                        'caloric_energy_from_cereals_roots_tubers',
                        'access_to_improved_sanitation', 'access_to_improved_water_sources',
                        'anemia_prevalence', 'obesity_prevalence', 'open_defecation',
                        'hiv_incidence', 'rail_lines_density', 'access_to_electricity',
                        'co2_emissions', 'unemployment_rate', 'total_labor_force',
                        'military_expenditure_share_gdp',
                        'proportion_of_seats_held_by_women_in_gov', 'political_stability']
l = 'prevalence_of_undernourishment'

# Without this I get a very annoying warning :-)
warnings.filterwarnings("ignore", category=UserWarning, module="matplotlib")

for f in features_of_interest:
    print("\nFeature: {}".format(f))
    df = pd.concat([train_dataset[f], train_dataset[l]], axis='columns')
    percent_NaN = df[f].isnull().mean().round(4)*100

    with warnings.catch_warnings():
        fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,4), tight_layout=True)

    df[f].plot(ax=axes[0], kind='kde', secondary_y=True)
    df[f].plot(ax=axes[0], kind='hist')
    axes[0].annotate("% NaN:\n  {:.2f}".format(percent_NaN),
                xy=(.75, .85), xycoords=axes[0].transAxes)
    axes[0].set_title("{}".format(f));
    df[f].plot(ax=axes[1], kind='box');
    axes[1].set_title("{}".format(f));
    df[f].plot(ax=axes[2], kind='box', logy=True)
    axes[2].set_title("log({})".format(f));
    plt.show()
    plt.close()
```

## Explore Prevalence of Undernourishment.

```python
f = 'prevalence_of_undernourishment'
df = pd.merge(features, labels, on='row_id', how='inner')

# Want to overlay multiple graphs hence subplots
fig, ax1 = plt.subplots(figsize=(16,9), frameon=True)
# Make sure second axis (graph) is twinned with first.
ax2 = ax1.twinx()
# Set Figure level properties. (Figure is the container?)
fig.subplots_adjust(top=0.90)
fig.suptitle('Histogram and density plot of PoU', fontsize=32, fontname='Calibri') # Add the text/suptitle to figure

# Use Pandas plot framework
df[f].plot(ax=ax1, kind='hist', xlim=(0, 70), color="C1")
df[f].plot(ax=ax2, kind='kde', xlim=(0, 70), color="C0")
# Tweak overall titles and styles
ax1.set_xlabel('Prevalence of Undernourishment',fontsize = 24, fontname='Calibri')
ax1.set_ylabel(ylabel="Frequency", fontsize = 20, fontname='Calibri')
ax2.set_ylabel(ylabel="Density", fontsize = 20, fontname='Calibri')

# Add mean and median lines, as well as a legend
mean_val = df['prevalence_of_undernourishment'].mean()
median_val = df['prevalence_of_undernourishment'].median()
ax1.axvline(label="Mean", x=mean_val, ymin=0.0, ymax=.9, color="C3", linestyle='--')
ax1.axvline(label="Median", x=median_val, ymin=0.0, ymax=.9, color="C9", linestyle='--')

# Add a legend
import matplotlib.patches as mpatches
mean_patch = mpatches.Patch(color='C3', label='Mean')
median_patch = mpatches.Patch(color='C9', label='Median')
plt.legend(handles=[mean_patch, median_patch], fontsize=20)

plt.savefig(fname='images/PoU_Histogram.png', bbox_inches='tight')
plt.show()
plt.close(fig)
```

## Calculate the summary statistics

```python
# Descriptive statistics
#pd.set_option('display.precision', 2)

sdf = pd.merge(features, labels, on='row_id', how='inner')

df = pd.DataFrame(columns=['Min', 'Max', 'Mean', 'Median', 'Std Dev', 'DCount', '% NaN'])
df['Min'] = sdf.iloc[:,3:48].min()
df['Max'] = sdf.iloc[:,3:48].max()
df['Mean'] = sdf.iloc[:,3:48].mean()
df['Median'] = sdf.iloc[:,3:48].median()
df['Std Dev'] = sdf.iloc[:,3:48].std()
result = []
for f in sdf.iloc[:,3:48].columns:
    result.append(pd.DataFrame(sdf[f].unique()).count()[0])
df['DCount'] = result
df['% NaN'] = sdf.iloc[:,3:48].isnull().mean().round(4)*100
df.to_csv('summary-statistics.csv')
print(df)

#pd.reset_option('display.precision')
```
