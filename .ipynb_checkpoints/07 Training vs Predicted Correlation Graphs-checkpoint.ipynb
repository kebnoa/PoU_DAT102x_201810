# 07 - Training vs Predicted Correlation Graph
Now that we've created a model, tuned it, and predicted the PoU we need to check how it compares to the actual values. We can only perform this test on the Training Dataset as we do not have the actual PoU values for the test.

## Initialise the styles for the workbooks

```python
# Initialise styles and packages we need
from IPython.core.display import HTML
def css_styling():
    styles = open("styles/custom.css", "r").read()
    return HTML(styles)
css_styling()
```

## Imports and classes used

```python
# All the imports used
import pandas as pd
# conventional way to import seaborn
import seaborn as sns; sns.set(style="ticks", color_codes=True)

import numpy as np

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patches as mpatches

from scipy.stats import pearsonr
from sklearn import preprocessing
import warnings

# allow plots to appear within the notebook
%matplotlib inline

print("Pandas version:  {}".format(pd.__version__))
print("Seaborn version: {}".format(sns.__version__))
print("Numpy version:   {}".format(np.__version__))
#print("Matplotlib version:  {}".format(matplotlib.__version__))

def corrfunc(x, y, **kws):
    (r, p) = pearsonr(x, y)
    ax = plt.gca()
    ax.annotate("pearsonr = {:.2f}; p = {:.3f}".format(r, p),
                xy=(.5, .95), xycoords=ax.transAxes)

def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

## Load the predicted and provided PoU values

```python
pred = 'predicted_prevalence_of_undernourishment'
act = 'prevalence_of_undernourishment'
df = pd.read_csv('actual_vs_predicted_PoU_20181025.csv')
df['abs_error'] = abs((df[pred] - df[act]))
print(df.head())
print(df.shape)
```

## Plot the correlation

```python
fig = plt.figure(figsize=(16,9))#, constrained_layout=True)
fig.suptitle("{}".format(act), fontsize=32, fontname='Calibri')
gs  = gridspec.GridSpec(1, 1)
ax0 = plt.subplot(gs[0])
sns.regplot(ax=ax0, x=df[pred], y=df[act], data=df, fit_reg=True)
(r, p) = pearsonr(df[act], df[pred])
ax0.annotate("pearsonr = {:.3f}; p = {:.3f}".format(r, p),
              xy=(.7, .95), xycoords=ax0.transAxes,
              fontsize=16)
ax0.set_xlabel(xlabel='')
ax0.set_ylabel(ylabel=pred, fontsize = 20, fontname='Calibri')
plt.savefig(fname='pred_vs_act_correlation.png', bbox_inches='tight')
plt.show()
plt.close(fig)
```

## Calculate the RMSE between the predicted and actual values

```python
c = rmse(df[pred], df[act])
print(c)
```

## Calculate the summary statistics of the actual and predicted values

```python
sdf = pd.concat([df[act], df[pred]], axis='columns')
#print(sdf)
dfs = pd.DataFrame(columns=['Min', 'Max', 'Mean', 'Median', 'Std Dev'])
dfs['Min'] = sdf.min()
dfs['Max'] = sdf.max()
dfs['Mean'] = sdf.mean()
dfs['Median'] = sdf.median()
dfs['Std Dev'] = sdf.std()
dfs.to_csv('summary_statistics_actual_vs_predicted_PoU_20181025.csv')
print(dfs)
```
