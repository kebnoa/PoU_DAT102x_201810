## Imports and classes used


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns

# allow plots to appear within the notebook
%matplotlib inline

from scipy.stats import pearsonr
def corrfunc(x, y, **kws):
    (r, p) = pearsonr(x, y)
    ax = plt.gca()
    ax.annotate("pearsonr = {:.2f}; p = {:.3f}".format(r, p),
                xy=(.5, .95), xycoords=ax.transAxes)
    
import warnings

# https://stackoverflow.com/questions/35042255/how-to-plot-multiple-seaborn-jointplot-in-subplot
class SeabornFig2Grid():

    def __init__(self, seaborngrid, fig,  subplot_spec):
        self.fig = fig
        self.sg = seaborngrid
        self.subplot = subplot_spec
        if isinstance(self.sg, sns.axisgrid.FacetGrid) or \
            isinstance(self.sg, sns.axisgrid.PairGrid):
            self._movegrid()
        elif isinstance(self.sg, sns.axisgrid.JointGrid):
            self._movejointgrid()
        self._finalize()

    def _movegrid(self):
        """ Move PairGrid or Facetgrid """
        self._resize()
        n = self.sg.axes.shape[0]
        m = self.sg.axes.shape[1]
        self.subgrid = gridspec.GridSpecFromSubplotSpec(n,m, subplot_spec=self.subplot)
        for i in range(n):
            for j in range(m):
                self._moveaxes(self.sg.axes[i,j], self.subgrid[i,j])

    def _movejointgrid(self):
        """ Move Jointgrid """
        h= self.sg.ax_joint.get_position().height
        h2= self.sg.ax_marg_x.get_position().height
        r = int(np.round(h/h2))
        self._resize()
        self.subgrid = gridspec.GridSpecFromSubplotSpec(r+1,r+1, subplot_spec=self.subplot)

        self._moveaxes(self.sg.ax_joint, self.subgrid[1:, :-1])
        self._moveaxes(self.sg.ax_marg_x, self.subgrid[0, :-1])
        self._moveaxes(self.sg.ax_marg_y, self.subgrid[1:, -1])

    def _moveaxes(self, ax, gs):
        #https://stackoverflow.com/a/46906599/4124317
        ax.remove()
        ax.figure=self.fig
        self.fig.axes.append(ax)
        self.fig.add_axes(ax)
        ax._subplotspec = gs
        ax.set_position(gs.get_position(self.fig))
        ax.set_subplotspec(gs)

    def _finalize(self):
        plt.close(self.sg.fig)
        self.fig.canvas.mpl_connect("resize_event", self._resize)
        self.fig.canvas.draw()

    def _resize(self, evt=None):
        self.sg.fig.set_size_inches(self.fig.get_size_inches())
```

## Read the files into Pandas

```python
training_values_filename = 'data/DAT102x_Predicting_Chronic_Hunger_-_Training_values.csv'
test_values_filename = 'data/DAT102x_Predicting_Chronic_Hunger_-_Test_values.csv'

training_values = pd.read_csv(training_values_filename)
test_values = pd.read_csv(test_values_filename)

print("Training values: {}".format(training_values.shape))
print("Test values:     {}".format(test_values.shape))
```

## Remove features as determined in Data Exploration

```python
features_to_drop = ['droughts_floods_extreme_temps',
                    'inequality_index',
                    'tax_revenue_share_gdp',
                    'adult_literacy_rate',
                    'school_enrollment_rate_female',
                    'school_enrollment_rate_total',
                    'hiv_incidence',
                    'rail_lines_density',
                    'open_defecation',
                    'rural_population',
                    'percentage_of_arable_land_equipped_for_irrigation',
                    'cereal_yield',
                    'anemia_prevalence',
                    'unemployment_rate',
                    'total_labor_force',
                    'military_expenditure_share_gdp',
                    'proportion_of_seats_held_by_women_in_gov',
                    'political_stability',
                    'fertility_rate',
                    'life_expectancy',
                    'trade_in_services',
                    'per_capita_food_production_variability',
                    'net_oda_received_percent_gni',
                    'net_oda_received_per_capita',
                    'cereal_import_dependency_ratio',
                    'imports_of_goods_and_services']
training_values.drop(features_to_drop, axis='columns', inplace=True)
test_values.drop(features_to_drop, axis='columns', inplace=True)

print("Training values: {}".format(training_values.shape))
print("Test values:     {}".format(test_values.shape))
```

## Remove countries as determined in Data Exploration

```python
countries_to_drop = ['0b6e276', 'be60011', '79c89fd', '93d74a6', 'f196e55', 'd3a0eeb']

# first need to find all the observations, i.e. row_ids
obs = training_values[training_values.country_code.isin(countries_to_drop)].row_id

# now go remove them from training
training_values.drop(obs, axis='rows', inplace=True)

print("Training values: {}".format(training_values.shape))
print("Test values:     {}".format(test_values.shape))
```

## Use Pandas interpolation to estimate missing values by country

```python
# For Training data:
features = training_values.columns
interpolated = pd.DataFrame(columns=features)
for country in training_values.country_code.unique():
    df = training_values[training_values['country_code'] == country].copy()
    df.sort_values(by='year', inplace=True)
    # Need to do interpolation forward and backwards seperately
    df.interpolate(limit_direction='forward', inplace=True)
    # In AzureML (Pandas 0.18.0) you must sort to make the backward changes happen.
    # Here (Pandas 0.23.1) you must NOT sort, else it doesn't work!
#    df.sort_values(by='year', ascending=False, inplace=True)
    df.interpolate(limit_direction='backward', inplace=True)
    interpolated = interpolated.append(df).sort_values(by='row_id')
interpolated_training_values = interpolated

# For Testing data:
features = test_values.columns
interpolated = pd.DataFrame(columns=features)
for country in test_values.country_code.unique():
    df = test_values[test_values['country_code'] == country].copy()
    df.sort_values(by='year', inplace=True)
    # Need to do interpolation forward and backwards seperately
    # Also, not sure if Azure or Pandas 0.18.0 but must sort in-between to make it work!
    df.interpolate(limit_direction='forward', inplace=True)
    # In AzureML (Pandas 0.18.0) you must sort to make the backward changes happen.
    # Here (Pandas 0.23.1) you must NOT sort, else it doesn't work!
#    df.sort_values(by='year', ascending=False, inplace=True)
    df.interpolate(limit_direction='backward', inplace=True)
    interpolated = interpolated.append(df).sort_values(by='row_id')
interpolated_test_values = interpolated

print("Training values: {}".format(interpolated_training_values.shape))
print("Test values:     {}".format(interpolated_test_values.shape))
```

## Confirm no remaining NaN (aka NULL) values

```python
row_ids = interpolated_training_values[interpolated_training_values.isnull().any(axis=1)].row_id
print("\nTraining observations with NaN values:")
for row in row_ids:
    print("{},".format(row))

row_ids = interpolated_test_values[interpolated_test_values.isnull().any(axis=1)].row_id
print("\nTest observations with NaN values:")
for row in row_ids:
    print("{},".format(row))
```

## Add the ratio_urban_population_total feature

```python
# Based on questions from DAT102x data exploration add the following feature and remove composite features
final_training_values = interpolated_training_values
final_training_values['ratio_urban_population_total'] = \
    final_training_values['urban_population'] / final_training_values['total_population']
final_training_values.drop(['urban_population','total_population'], axis='columns', inplace=True)

final_test_values = interpolated_test_values
final_test_values['ratio_urban_population_total'] = \
    final_test_values['urban_population'] / final_test_values['total_population']
final_test_values.drop(['urban_population','total_population'], axis='columns', inplace=True)

print("Training values: {}".format(final_training_values.shape))
print("Test values:     {}".format(final_test_values.shape))
```

## Save the cleanded data sets for re-use

```python
# Write out to CSV files so we can use in next phase
final_training_values_filename = 'data/DAT102x_Predicting_Chronic_Hunger_-_Clean_Training_values.csv'
final_test_values_filename = 'data/DAT102x_Predicting_Chronic_Hunger_-_Clean_Test_values.csv'

final_training_values.to_csv(final_training_values_filename, index=False)
final_test_values.to_csv(final_test_values_filename, index=False)
```

## Create correlation graphs for reporting purposes

```python
final_training_values_filename = 'data/DAT102x_Predicting_Chronic_Hunger_-_Clean_Training_values.csv'
training_labels_filename = 'data/DAT102x_Predicting_Chronic_Hunger_-_Training_labels.csv'
training_values = pd.read_csv(final_training_values_filename)
training_labels = pd.read_csv(training_labels_filename)
# Match up training lables to training features
tempDF = pd.merge(training_values, training_labels, on='row_id', how='inner')
print(tempDF.shape)
```

```python
features_of_interest = ['population_growth', 'avg_value_of_food_production',
                        'gross_domestic_product_per_capita_ppp',
                        'avg_supply_of_protein_of_animal_origin',
                        'caloric_energy_from_cereals_roots_tubers',
                        'access_to_improved_sanitation', 'access_to_improved_water_sources',
                        'obesity_prevalence', 'access_to_electricity',
                        'ratio_urban_population_total']
l = 'prevalence_of_undernourishment'

sns.set(style="ticks")

for f in features_of_interest:
    df = pd.concat([tempDF[f], tempDF[l]], axis='columns')
    g0 = sns.pairplot(df, x_vars=[l], y_vars=[f], height=5, aspect=0.8, kind='reg')
    g0.map(corrfunc)
    g1 = sns.jointplot(l, f, data=df, height=5, ratio=3, kind = 'hex')

    fig = plt.figure(figsize=(12,8))
    gs = gridspec.GridSpec(1,2)
    
    mg0 = SeabornFig2Grid(g0, fig, gs[0])
    mg1 = SeabornFig2Grid(g1, fig, gs[1])

    gs.tight_layout(fig)
    plt.show()
    plt.close()
```
